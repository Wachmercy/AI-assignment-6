```python
# ==========================================
# Lightweight Image Classification with CNN
# ==========================================

# Step 1: Install TensorFlow (if not already installed)
!pip install -q tensorflow

# Step 2: Import required libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import cifar10
import matplotlib.pyplot as plt
import numpy as np

# Step 3: Load CIFAR-10 dataset (10 classes, 32x32 images)
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

# Normalize pixel values to range [0,1] for faster convergence
x_train, x_test = x_train / 255.0, x_test / 255.0

# Print dataset shapes for verification
print("Training set:", x_train.shape, y_train.shape)
print("Test set:", x_test.shape, y_test.shape)

# Step 4: Define a lightweight Convolutional Neural Network (CNN)
model = models.Sequential([
    # First convolutional layer with 32 filters, 3x3 kernel
    layers.Conv2D(32, (3,3), activation='relu', input_shape=(32,32,3)),
    # MaxPooling reduces spatial size (downsampling)
    layers.MaxPooling2D((2,2)),

    # Second convolutional layer with 64 filters
    layers.Conv2D(64, (3,3), activation='relu'),
    layers.MaxPooling2D((2,2)),

    # Flatten 2D feature maps into 1D vector for Dense layers
    layers.Flatten(),

    # Fully connected layer with 64 neurons
    layers.Dense(64, activation='relu'),

    # Output layer with 10 neurons (one per class) using softmax
    layers.Dense(10, activation='softmax')
])

# Step 5: Compile the model
# Optimizer: Adam (adaptive learning rate)
# Loss: SparseCategoricalCrossentropy (for integer labels)
# Metrics: Accuracy
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Print model summary to see architecture
model.summary()

# Step 6: Train the model
# epochs: number of training cycles
# batch_size: number of samples per gradient update
# validation_data: test set for monitoring performance
history = model.fit(x_train, y_train,
                    epochs=10,
                    batch_size=64,
                    validation_data=(x_test, y_test))

# Step 7: Evaluate model on test set
test_loss, test_acc = model.evaluate(x_test, y_test)
print("Test accuracy:", test_acc)

# Step 8: Make predictions on some test images
class_names = ['airplane','automobile','bird','cat','deer',
               'dog','frog','horse','ship','truck']

predictions = model.predict(x_test[:5])

# Step 9: Visualize predictions
for i in range(5):
    plt.imshow(x_test[i])
    plt.title(f"Predicted: {class_names[np.argmax(predictions[i])]}, Actual: {class_names[y_test[i][0]]}")
    plt.axis('off')
    plt.show()


```

```python
# ==========================================
# TensorFlow Lite Conversion and Testing
# ==========================================

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Convert the trained model to TensorFlow Lite format
# This creates a lightweight model suitable for on-device inference
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()

# Step 2: Save the TFLite model to a file
tflite_model_file = "lightweight_cnn.tflite"
with open(tflite_model_file, "wb") as f:
    f.write(tflite_model)

print(f"TFLite model saved as {tflite_model_file}")

# Step 3: Load the TFLite model using the Interpreter
interpreter = tf.lite.Interpreter(model_path=tflite_model_file)
interpreter.allocate_tensors()

# Step 4: Get input and output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print("Input details:", input_details)
print("Output details:", output_details)

# Step 5: Run inference on a few sample test images
# Take first 5 images from CIFAR-10 test set
sample_images = x_test[:5]
sample_labels = y_test[:5]

# Class names
class_names = ['airplane','automobile','bird','cat','deer',
               'dog','frog','horse','ship','truck']

for i, img in enumerate(sample_images):
    # TFLite interpreter requires input shape (1, height, width, channels)
    input_data = np.expand_dims(img, axis=0).astype(np.float32)

    # Set the tensor to point to the input data
    interpreter.set_tensor(input_details[0]['index'], input_data)

    # Run inference
    interpreter.invoke()

    # Get prediction results
    output_data = interpreter.get_tensor(output_details[0]['index'])
    predicted_class = np.argmax(output_data)

    # Display the image and prediction
    plt.imshow(img)
    plt.title(f"Predicted: {class_names[predicted_class]}, Actual: {class_names[sample_labels[i][0]]}")
    plt.axis('off')
    plt.show()
```

## Outline Deployment Steps

### Deployment of `.tflite` Models for On-Device Inference

1.  **General Concept and Benefits of `.tflite` Deployment:**

    TensorFlow Lite (TFLite) is a lightweight solution for deploying machine learning models on edge devices, such as mobile phones (Android, iOS), embedded Linux devices, and microcontrollers. The primary benefit of using `.tflite` models is their optimized format, which enables low-latency inference with a smaller memory footprint and reduced power consumption, making them ideal for on-device AI applications where cloud connectivity might be intermittent or not feasible. This allows applications to run intelligent features offline, improve user privacy by processing data locally, and reduce operational costs.

2.  **Loading the `.tflite` Model:**

    The first step in deploying a `.tflite` model is to load it onto the target device using the TensorFlow Lite Interpreter. The interpreter is a core component of the TFLite runtime that executes the model. The process typically involves:

    *   **Including the TFLite runtime library:** Depending on the platform (e.g., Android, iOS, C++), the appropriate TFLite runtime library needs to be integrated into the application.
    *   **Loading the `.tflite` file:** The `.tflite` model file (e.g., `lightweight_cnn.tflite`) is loaded into memory. In Python, this is done via `tf.lite.Interpreter(model_path='path/to/model.tflite')`. For other languages, similar APIs exist to create an interpreter instance from the model file or a byte array.
    *   **Allocating Tensors:** After loading, `interpreter.allocate_tensors()` is called to allocate memory for the model's input, output, and intermediate tensors.
    *   **Getting Input/Output Details:** Retrieve `input_details` and `output_details` to understand the expected input shape, data type, and the structure of the output.

3.  **Preparing Input Data:**

    Before running inference, the input data (e.g., an image from the device's camera) must be preprocessed to match the format expected by the model. This typically involves:

    *   **Resizing:** Images are often resized to the model's input dimensions (e.g., 32x32 for our CIFAR-10 model).
    *   **Normalization:** Pixel values are scaled to the range the model was trained on (e.g., [0, 1] by dividing by 255.0).
    *   **Data Type Conversion:** Ensuring the input data has the correct data type (e.g., `float32`).
    *   **Dimension Expansion:** Adding a batch dimension if the model expects one (e.g., converting `(height, width, channels)` to `(1, height, width, channels)`).

4.  **Running Inference with the TFLite Interpreter:**

    Once the input data is prepared, inference can be performed:

    *   **Set Input Tensor:** The preprocessed input data is copied into the model's input tensor using `interpreter.set_tensor(input_details[0]['index'], input_data)`.
    *   **Invoke Interpreter:** The model's computation graph is executed by calling `interpreter.invoke()`.
    *   **Get Output Tensor:** The prediction results are retrieved from the output tensor using `interpreter.get_tensor(output_details[0]['index'])`.

5.  **Interpreting the Output:**

    The raw output from the model (e.g., a probability distribution over classes) needs to be interpreted into a human-readable format:

    *   **Post-processing:** For classification tasks, `np.argmax()` is commonly used to find the index of the highest probability, which corresponds to the predicted class label.
    *   **Mapping to Class Names:** The class index is then mapped to its corresponding class name (e.g., from `class_names` list) to provide a meaningful prediction.

These steps form the core workflow for integrating a TFLite model into various applications for efficient on-device inference.
